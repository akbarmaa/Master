**Q : Lakukan analisis PMF pada kolom Attrition_Flag**

```python
# PMF
data['Attrition_Flag'].value_counts(normalize=True)
```

**Q: Pisahkan data Existing Customer and Attrited customer pada kolom Attrition_Flag**

```python
plt.figure(figsize=(10,7))
plt.title('Pemisahan')
sns.countplot(x='Attrition_Flag',hue='Attrition_Flag',data=data)
```

**Q: Lakukan analisis distribusi pada kolom Total_Trans_Amt dan Total_Trans_Ct serta tentukan skewnessnya**

```python
# Distribution & Skewness
sns.histplot(data=data, x='Total_Trans_Amt', kde=True)
print("Skewness: ", data['Total_Trans_Amt'].skew())
plt.show()
```

```python
# Distribution & Skewness
sns.histplot(data=data, x='Total_Trans_Ct', kde=True)
print("Skewness: ", data['Total_Trans_Ct'].skew())
plt.show()
```

**Q: Lakukan analisis hubungan/korelasi antara kolom Total_Trans_Amt dan Total_Trans_Ct dan jelaskan apakah memiliki positive, negatif atau netral**

```python
plt.figure(figsize=(22,7))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

**Answer**

Berdasarkan Tabel korelasi Heatmap, Total_Trans_Amt dan Total_Trans_Ct berkorelasi positif sebesar 0,81% dimana ketika terjadi peningkatan value dari kolom Total_Trans_Amt, kolom Total_Trans_Ct juga akan mengalami peningkatan.
