**Import Libraries and Data**

```python
import pandas as pd
import numpy as np
from empiricaldist import Pmf, Cdf

import io
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```python
data = pd.read_csv(io.BytesIO(uploaded['ParisHousingClass.csv']))
data.head()
```

```python
data.isna().sum()
```

```python
data.hist(figsize=(20,15))
plt.show()
```

```python
data.category.value_counts().plot(kind='bar')
print(data.category.value_counts())
```

```python
plt.figure(figsize=(10,7))
plt.title('Pool')
sns.countplot(x='hasPool',hue='category',data=data)
```

```python
plt.figure(figsize=(10,7))
plt.title('Storage Room')
sns.countplot(x='hasStorageRoom',hue='category',data=data)
```

```python
plt.figure(figsize=(10,7))
plt.title('Kategori Bangunan')
sns.countplot(x='isNewBuilt',hue='category',data=data)
```

```python
pd.crosstab(data['category'], data['hasStormProtector'], margins=True).style.background_gradient(vmin=0)
```

```python
pd.crosstab(data['category'], data['hasGuestRoom'], margins=True).style.background_gradient(vmin=0)
```

```python
sns.boxplot(data.price)
```
